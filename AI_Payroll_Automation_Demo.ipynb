```python
import pandas as pd

data = {
    "Employee": ["John", "Sarah", "Mike", "Lisa"],
    "Hours": [40, 45, 85, 38],
    "Rate": [25, 30, 28, 22],
    "Bonus": [100, 0, 200, 50]
}

df = pd.DataFrame(data)

df["Gross Pay"] = (
    df["Hours"] * df["Rate"]
    + df["Bonus"]
)

df["Warning"] = df["Hours"].apply(
    lambda x: "Check Hours" if x > 80 else ""
)

print("PAYROLL SUMMARY")
print(df)
```

```python
total_payroll = df["Gross Pay"].sum()

print("===================================")
print(" CPA PAYROLL AUTOMATION REPORT")
print("===================================")

print(f"Total Employees: {len(df)}")
print(f"Total Payroll: ${total_payroll:,.2f}")

print("\nEmployees Requiring Review:")

review_found = False

for index, row in df.iterrows():
    if row["Hours"] > 80:
        print(f"- {row['Employee']} worked {row['Hours']} hours")
        review_found = True

if not review_found:
    print("No issues found.")
```

```python
average_hours = df["Hours"].mean()

for index, row in df.iterrows():
    if row["Hours"] > average_hours * 1.5:
        print(
            f"AI Insight: {row['Employee']} worked significantly more hours "
            f"than the team average ({average_hours:.1f} hours)."
        )
```
